```python
df = pd.read_csv(r"C:\Users\Lenovo\數據分析\天下專案\10- Key Metrics.csv")
likes = pd.read_csv(r"C:\Users\Lenovo\數據分析\天下專案\like.csv")
```

```python
likes2 = pd.read_csv(r"C:\Users\Lenovo\數據分析\天下專案\like2.csv")
```

```python
likes_clean = likes[['已發佈','永久連結','like','comment','share']]
likes2_clean = likes2[['已發佈','永久連結','like','comment','share']]
likes2_clean['month'] = likes2_clean['已發佈'].apply(lambda x: x[:2])
#取原本 "已發佈" 變數的前2個字 
likes2_clean = likes2_clean[likes2_clean['month'] == '10']
```

```python
all_like = likes_clean.append(likes2_clean)
#all_like.to_excel(r"C:\Users\Lenovo\數據分析\天下專案\alllike.xlsx", encoding="utf8", index = False)
all_like= all_like.reset_index(drop=True)
```

```python
browser = webdriver.Chrome(executable_path = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
proxy = {
        'https': 'https://91.206.30.218:3128'
    }
header = {
        'User-agent': 'Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1)',
    }
info = pd.DataFrame(columns = ['標題'])
for i in tqdm_notebook(range(1,len(all_like['永久連結']))):
    browser.get(all_like['永久連結'][i])
    soup7 = BeautifulSoup(browser.page_source, "html.parser") 
    try:
        com_info2 = pd.DataFrame([soup7.find("a", class_="_52c6")["aria-label"]], columns = ['標題'])
        info = pd.concat([info, com_info2], axis = 0)
        #用一開始建立的info來跟後面的com_info2使用 concat 合併 axis=0 為直向合併
        info = info.reset_index(drop = True)
    except:
        blocks = soup7.find('title', id = "pageTitle")
        com_info2 = pd.DataFrame([blocks.text], columns = ['標題'])
        info = pd.concat([info, com_info2], axis = 0)
        #用一開始建立的info來跟後面的com_info2使用 concat 合併 axis=0 為直向合併
        info = info.reset_index(drop = True)
```
